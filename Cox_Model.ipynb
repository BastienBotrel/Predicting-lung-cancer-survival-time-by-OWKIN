### **Features Selection**

**Univariate Feature Selection**

A first selection of the variables is done by performing univariate analyses. Variables which are not associated with overall survival at 15% threshold will not be included in a multivariate analysis.

```
from lifelines import CoxPHFitter
cph = CoxPHFitter()
```

```
# Dataset containing all features as well as survival time and censorhsip variable
x_train_model=pd.merge(x_train, y_train, how='inner', left_index=True, right_index=True, sort=True)
```

```
# Univariate analysis
for variable in x_train.columns:
  x_train_selection=x_train_model.loc[:,[f'{variable}','SurvivalTime','Event']]
  cph.fit(x_train_selection, duration_col='SurvivalTime', event_col='Event')
  cph.print_summary()
```

The following 8 variables will not be included in the multivariate analysis: *original_shape_Compactness1, original_shape_Compactness2, original_shape_SphericalDisproportion, original_shape_Sphericity, original_firstorder_Energy, original_firstorder_Minimum, original_glcm_ClusterProminence, original_glrlm_LongRunLowGrayLevelEmphasis*

**Multivariate Analysis** with selected variables

```
x_train_model_ = x_train_model.drop(columns=["original_shape_Compactness1", "original_shape_Compactness2", "original_shape_SphericalDisproportion", 
                                   "original_shape_Sphericity", "original_firstorder_Energy", "original_firstorder_Minimum", "original_glcm_ClusterProminence", 
                                    "original_glrlm_LongRunLowGrayLevelEmphasis"])

cph.fit(x_train_model_, duration_col='SurvivalTime', event_col='Event')
cph.print_summary()
```

We have the following warning message: "*Newton-Rhaphson failed to converge sufficiently*". It may be due to ***multicollinearity*** which occurs when some features are highly linearly related. 
Here, the Newton-Raphson's method is used to maximized the Cox partial likelihood.

**Correlation matrix** of continuous features to have insight into which of the features are higly correlated

```
x_train_model_numeric = x_train_model_.drop(columns=['SourceDataset','SurvivalTime','Event','Nstage_1','Nstage_2','Nstage_3','Tstage_2','Tstage_3','Tstage_3','Tstage_4',
                                                     'Histology_cat_1','Histology_cat_2', 'Histology_cat_3'])

# Compute the correlation matrix
corr = x_train_model_numeric.corr()
corr
```

```
#Set up the matplotlib figure
f, ax = plt.subplots(figsize=(30, 15))

#Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

#Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 20, as_cmap=True)
sns.heatmap(corr, cmap=cmap, mask=mask, vmin=0, vmax=1)
```

We can notice that some features are highly correlated.

To assess the multicollinearity, we can compute the **Variance Inflation Factor** (VIF). Its measures how much the variance of an estimated regression coefficient is increased because of collinearity. When VIF is high (>10), there is high multicollinearity and instability of the
coefficients.

```
# Compute VIF
from pandas.testing import assert_frame_equal
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["variables"] = x_train_model_numeric.columns
vif["VIF"] = [variance_inflation_factor(x_train_model_numeric.values, i) for i in range(x_train_model_numeric.shape[1])]
vif
```

For several variables, the VIF is > 100, meaning that more of 99% of their variance is explained by the other variables.

Afer analysing the pairwise correlations and the VIFs, I decided to remove the following variables from the model: *original_glcm_Id,original_glcm_Idm,original_firstorder_Range, original_glcm_JointEntropy, original_glrlm_HighGrayLevelRunEmphasis, original_firstorder_Entropy,original_firstorder_Uniformity, original_firstorder_Variance,original_glcm_Autocorrelation, original_glcm_DifferenceAverage,original_glcm_DifferenceEntropy, original_glcm_Idn,original_firstorder_MeanAbsoluteDeviation" original_firstorder_RootMeanSquared,original_glrlm_RunPercentage, original_glrlm_ShortRunLowGrayLevelEmphasis, original_glcm_SumEntropy,original_glcm_SumAverage original_firstorder_StandardDeviation, original_firstorder_Mean,original_glcm_JointEnergy,original_shape_VoxelVolume,original_glrlm_ShortRunEmphasis, original_glrlm_LongRunHighGrayLevelEmphasis, original_shape_SurfaceArea,original_glcm_Contrast, original_glcm_Imc1, original_glrlm_LongRunEmphasis.*

Pairwise correlations between the remaining variables are <0.85 and VIFs are < 10.

```
x_train_model_2 = x_train_model_.drop(columns=["original_glcm_Id","original_glcm_Idm","original_firstorder_Range", "original_glcm_JointEntropy",
                                      "original_glrlm_HighGrayLevelRunEmphasis","original_firstorder_Entropy","original_firstorder_Uniformity",
                                      "original_firstorder_Variance","original_glcm_Autocorrelation","original_glcm_DifferenceAverage","original_glcm_DifferenceEntropy",
                                      "original_glcm_Idn","original_firstorder_MeanAbsoluteDeviation","original_firstorder_RootMeanSquared","original_glrlm_RunPercentage",
                                      "original_glrlm_ShortRunLowGrayLevelEmphasis","original_glcm_SumEntropy","original_glcm_SumAverage","original_firstorder_StandardDeviation",
                                      "original_firstorder_Mean","original_glcm_JointEnergy","original_shape_VoxelVolume","original_glrlm_ShortRunEmphasis",
                                      "original_glrlm_LongRunHighGrayLevelEmphasis","original_shape_SurfaceArea","original_glcm_Contrast","original_glcm_Imc1",
                                      "original_glrlm_LongRunEmphasis"])
```

### **Sequential Backward Selection**

Selected features are included in a multivariate model and the less significant are removed one by one, to keep only features significant at 5% threshold.

```
#This model contains 17 radiomics features and 5 clinical variables.
cph.fit(x_train_model_2, duration_col='SurvivalTime', event_col='Event')
cph.print_summary()
```

**Final Model**

The following variables have been removed in the following order: 1)original_glcm_ClusterShade, 2)original_glrlm_RunLengthNonUniformity, 3) original_shape_Maximum3DDiameter, 4)original_glcm_Imc2, 5)original_glrlm_LowGrayLevelRunEmphasis, 6)original_glrlm_ShortRunHighGrayLevelEmphasis, 7)original_glcm_Idmn,
8)original_firstorder_Kurtosis,9) Tstage_2, Tstage_3, Tstage_4, 
9)original_glcm_Correlation, 10)original_firstorder_Skewness, 11)original_glcm_InverseVariance, 12)original_shape_SurfaceVolumeRatio, 13)original_glcm_MaximumProbability,14)Histology_cat_1, Histology_cat_2,Histology_cat_3, 15)original_firstorder_Maximum. 

```
x_train_model_final = x_train_model_2.drop(columns=["original_glcm_ClusterShade","original_glrlm_RunLengthNonUniformity","original_shape_Maximum3DDiameter","original_glcm_Imc2",
                                                    "original_glrlm_LowGrayLevelRunEmphasis","original_glrlm_ShortRunHighGrayLevelEmphasis","original_glcm_Idmn",
                                                    "original_firstorder_Kurtosis","Tstage_2","Tstage_3","Tstage_4","original_glcm_Correlation","original_firstorder_Skewness",
                                                    "original_glcm_InverseVariance","original_shape_SurfaceVolumeRatio","original_glcm_MaximumProbability","Histology_cat_1",
                                                    "Histology_cat_2","Histology_cat_3","original_firstorder_Maximum"])

cph.fit(x_train_model_final, duration_col='SurvivalTime', event_col='Event')
cph.print_summary()
```

The final model contains 6 variables: *median voxel intensity, cluster tendency, gray level non-uniformity, age, dataset of origin and the N-tumoral stage*.

By default, the **Breslow's** method is used to estimate the baseline hazard function and the **Efron's** method is used handle situations in which there are ties in the survival time data. It is suitable to our data because we have very few tied event times.

```
from lifelines.utils import concordance_index
concordance=concordance_index(x_train_model_final['SurvivalTime'], -cph.predict_partial_hazard(x_train_model_final), x_train_model_final['Event']) 
print(concordance)
```

The **concordance index** on the train data is equal to 0.71
